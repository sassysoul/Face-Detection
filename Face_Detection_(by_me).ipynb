```python
#importing-libraries
import cv2 #opencv
import matplotlib.pyplot as plt
```

```python
img = plt.imread('AirBrush_20200214182958(1).jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
model = cv2.CascadeClassifier('haarcascade_frontalcatface.xml')
```

```python
faces = model.detectMultiScale(img)
```

```python
faces
```

```python
faces.shape
```

```python
for face in faces:
  x,y,w,h = face
  img = cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),20)
```

```python
plt.imshow(img)
```
